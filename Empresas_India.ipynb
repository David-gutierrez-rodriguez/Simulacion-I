```python
def razones_sociales(azar, tipos_empresas, actividades_economicas, nombres_comunes):
    combinaciones = product(("a", "b", "c", "d"), actividades_economicas, nombres_comunes, tipos_empresas)

    razones_sociales = []
    for _, actividad, nombre, tipo in combinaciones:
        razon = f"{actividad} {nombre} {tipo}"
        razones_sociales.append(razon)

    razones_sociales = rd.sample(razones_sociales, 1000)

    for i in range(1000):
        razones_sociales[i] = azar[i] + " " + razones_sociales[i]

    return razones_sociales

def sector(act):
    if act == "Agronomy":
        return "Primario"
    elif act in actividades_economicas[1:6]:
        return "Secundario"
    else:
        return "Terciario"

def numeros(acts):
  for act in acts:
    empleados = []
    if act == "Manufacture":
      for i in range(1000):
        e = int(np.random.normal(200, 100))
        while e <= 0:
          e = int(np.random.normal(200, 100))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(300, 150), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(250, 100), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(20, 5), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Farmaceutic":
      for i in range(1000):
        e = int(np.random.normal(300, 50))
        while e <= 0:
          e = int(np.random.normal(300, 50))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(400, 200), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(300,150), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(50, 7), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Auto":
      for i in range(1000):
        e = int(np.random.normal(500, 200))
        while e <= 0:
          e = int(np.random.normal(500, 200))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(600, 300), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(500, 200), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(10.6, 3), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Fabric":
      for i in range(1000):
        e = int(np.random.normal(800, 400))
        while e <= 0:
          e = int(np.random.normal(800, 400))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(250, 100), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(200, 100), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(30, 8), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Agronomy":
      for i in range(1000):
        e = int(np.random.normal(300, 150))
        while e <= 0:
          e = int(np.random.normal(300, 150))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(200, 100), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(150, 80), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(27, 6), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Finance":
      for i in range(1000):
        e = int(np.random.normal(200, 100))
        while e <= 0:
          e = int(np.random.normal(200, 100))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(350, 150), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(400, 200), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(80, 10), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Inmobiliarie":
      for i in range(1000):
        e = int(np.random.normal(100, 50))
        while e <= 0:
          e = int(np.random.normal(100, 50))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(300, 150), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(400, 200), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(44, 8), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Education":
      for i in range(1000):
        e = int(np.random.normal(150, 80))
        while e <= 0:
          e = int(np.random.normal(150, 80))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(150, 80), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(100, 50), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(56, 8), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Turism":
      for i in range(1000):
        e = int(np.random.normal(120, 40))
        while e <= 0:
          e = int(np.random.normal(120, 40))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(75, 25), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(50, 20), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(25, 3), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "Energy":
      for i in range(1000):
        e = int(np.random.normal(150, 80))
        while e <= 0:
          e = int(np.random.normal(150, 80))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(200, 100), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(180, 80), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(35, 7), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
    elif act == "IT&Software":
      for i in range(1000):
        e = int(np.random.normal(1000, 300))
        while e <= 0:
          e = int(np.random.normal(1000, 300))
        empleados.append(e)
      ventas_anuales = [round(np.random.normal(500, 200), 2) for _ in range(1000)]
      activos_totales = [round(np.random.normal(300, 100), 2) for _ in range(1000)]
      utilidad_bruta = [round(np.random.normal(36, 5), 2) for _ in range(1000)]
      return empleados, ventas_anuales, activos_totales, utilidad_bruta
```

```python
razones = razones_sociales(azar, tipos_empresas, actividades_economicas, nombres_comunes)
actividad_economica = [act.split(" ")[1] for act in razones]
sector_economico = [sector(act) for act in actividad_economica]
ciudad = [rd.choice(list(direcciones.keys())) for _ in range(1000)]
estado = [rd.choice(list(direcciones[ciudad].keys())) for ciudad in ciudad]
calle = [rd.choice(direcciones[ciudad[i]][estado[i]]) for i in range(1000)]
direccion = [calle[i] + ", " + estado[i] + ", " + ciudad[i] + ", India" for i in range(1000)]
empleados, ventas_anuales, activos_totales, utilidad_bruta = numeros(actividad_economica)

# ------------------------------------------
# Extras
# ------------------------------------------
productidad_por_empleado = [round(ventas_anuales[i]/empleados[i], 2) for i in range(1000)]
utilidad_por_empleado = [round(utilidad_bruta[i]/empleados[i], 2) for i in range(1000)]
activos_por_empleado = [round(activos_totales[i]/empleados[i], 2) for i in range(1000)]
costos_de_venta_por_empleado = [round((ventas_anuales[i] - utilidad_bruta[i])/empleados[i], 2) for i in range(1000)]
rotacion_de_activos = [round(ventas_anuales[i]/activos_totales[i], 2) for i in range(1000)]

# ------------------------------------------
# Data Frame
# ------------------------------------------
df = pd.DataFrame({
    "Razón Social": razones,
    "Actividad Económica": actividad_economica,
    "Dirección" : direccion,
    "Ciudad" : ciudad,
    "Estado" : estado,
    "Sector Económico" : sector_economico,
    "Empleados" : empleados,
    "Ventas Netas Anuales (₹M)" : ventas_anuales,
    "Activos Totales (₹M)" : activos_totales,
    "Utilidad Bruta (₹M)" : utilidad_bruta,
    "Productividad por Empleado (₹M)" : productidad_por_empleado,
    "Utilidad por Empleado (₹M)" : utilidad_por_empleado,
    "Activos por Empleado (₹M)" : activos_por_empleado,
    "Costos de Venta por Empleado (₹M)" : costos_de_venta_por_empleado,
    "Rotación de Activos" : rotacion_de_activos
})
display(df)
```

# Análisis comparativo

```python
# -----------------------------------------------------------------------
# Estados con peor desempeño económico
# -----------------------------------------------------------------------
estado_utilidad = df.groupby("Estado")["Utilidad Bruta (₹M)"].mean().sort_values()
peores_estados = estado_utilidad.head(2).index.tolist()

df_peores_estados = df[df["Estado"].isin(peores_estados)]

print("\nEstados con peor desempeño económico (utilidad bruta promedio):")
print(estado_utilidad.head(2))

# -----------------------------------------------------------------------
# Gráficas
# -----------------------------------------------------------------------
df_peores_estados.boxplot(column="Utilidad Bruta (₹M)", by="Estado", grid=False)
plt.title("Utilidad Bruta por Estado")
plt.suptitle("")
plt.ylabel("Utilidad Bruta (₹M)")
plt.show()

for estado in peores_estados:
    datos = df[df["Estado"] == estado]["Utilidad Bruta (₹M)"]
    plt.hist(datos, bins=15, alpha=0.5, label=estado)
plt.title("Histograma de Utilidad Bruta")
plt.xlabel("Utilidad Bruta (₹M)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------------------------------------
# Sectores con mejor desempeño económico
# -----------------------------------------------------------------------
sector_utilidad = df.groupby("Sector Económico")["Utilidad Bruta (₹M)"].mean().sort_values(ascending=False)
mejores_sectores = sector_utilidad.head(2).index.tolist()

df_mejores_sectores = df[df["Sector Económico"].isin(mejores_sectores)]

print("\nSectores con mejor desempeño económico (utilidad bruta promedio):")
print(sector_utilidad.head(2))

# -----------------------------------------------------------------------
# Gráficas
# -----------------------------------------------------------------------
df_mejores_sectores.boxplot(column="Utilidad Bruta (₹M)", by="Sector Económico", grid=False)
plt.title("Boxplot: Utilidad Bruta por Sector")
plt.suptitle("")
plt.ylabel("Utilidad Bruta (₹M)")
plt.show()

for sector in mejores_sectores:
    datos = df[df["Sector Económico"] == sector]["Utilidad Bruta (₹M)"]
    plt.hist(datos, bins=15, alpha=0.5, label=sector)
plt.title("Histograma de Utilidad Bruta")
plt.xlabel("Utilidad Bruta (₹M)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# -----------------------------------------------------------------------
# Ciudad con mejor vs peor desempeño económico
# -----------------------------------------------------------------------
ciudad_utilidad = df.groupby("Ciudad")["Utilidad Bruta (₹M)"].mean().sort_values()
peor_ciudad = ciudad_utilidad.index[0]
mejor_ciudad = ciudad_utilidad.index[-1]

df_ciudades_extremas = df[df["Ciudad"].isin([peor_ciudad, mejor_ciudad])]

print("\nCiudades con mayor y menor utilidad bruta promedio:")
print(ciudad_utilidad[[peor_ciudad, mejor_ciudad]])

# Gráficas
df_ciudades_extremas.boxplot(column="Utilidad Bruta (₹M)", by="Ciudad", grid=False)
plt.title("Boxplot: Utilidad Bruta por Ciudad")
plt.suptitle("")
plt.ylabel("Utilidad Bruta (₹M)")
plt.show()

for ciudad in [peor_ciudad, mejor_ciudad]:
    datos = df[df["Ciudad"] == ciudad]["Utilidad Bruta (₹M)"]
    plt.hist(datos, bins=15, alpha=0.5, label=ciudad)
plt.title("Histograma de Utilidad Bruta")
plt.xlabel("Utilidad Bruta (₹M)")
plt.ylabel("Frecuencia")
plt.legend()

plt.tight_layout()
plt.show()
```

# Tabla pivotada

```python
# VARIABLES A USAR (últimas 5)
variables = [
    "Productividad por Empleado (₹M)",
    "Utilidad por Empleado (₹M)",
    "Activos por Empleado (₹M)",
]

# ------------------------------------------
# Tabla pivotada con 1 variable
# ------------------------------------------
pivot1 = df.pivot_table(values=variables[0], index="Estado", aggfunc="mean")
print("\nTabla pivote con 1 variable (Productividad por Empleado):")
print(pivot1.round(2))

# ------------------------------------------
# Tabla pivotada con 2 variables
# ------------------------------------------
pivot2 = df.pivot_table(values=variables[0:2], index="Estado", aggfunc="mean")
print("\nTabla pivote con 2 variables (Productividad por Empleado y Utilidad por Empleado):")
print(pivot2.round(2))

# ------------------------------------------
# Tabla pivotada con 3 variables
# ------------------------------------------
pivot3 = df.pivot_table(values=variables, index="Estado", aggfunc="mean")
print("\nTabla pivot con 3 variables (Productividad por Empleado, Utilidad por Empleado y Activos por Empleado):")
print(pivot3.round(2))
```
