### Método de Montecarlo Crudo (media muestral) para cálculo de integrales

$$I = \int_a^b g(x) dx = \int_a^b w(x)f(x) dx = E[g(x)] \approx \frac{1}{N}\sum_{i = 1}^N g(x_i)$$

$$w(x) = g(x)(b - a)$$

$$f(x) = \frac{1}{b - a}$$

```python
import numpy as np
from random import random
import matplotlib.pyplot as plt
```

```python
# Función a la que se le calculará la integral
def g(x):
  return np.sqrt(np.atan(x))

# Método de Montecarlo para calcular la integral en el intervalo (a, b)
def integral(a, b, N):
  I = 0
  for i in range(N):
    I += g(random()*(b-a)+a)
  return I/N
```

```python
# Ejecución del cálculo de l integral 1000 veces
x = []
for i in range(1, 1000):
  x.append(integral(0, 1, i*10))
```

```python
# Valor real aproximado a 8 decimales

from statistics import stdev, mean

x0 = 0.62982334

# Gráfica de convergencia del método
plt.plot(x)
plt.xlabel('Número de iteraciones')
plt.ylabel('Valor de la integral')
plt.axhline(y = x0, color='r', linestyle='-')
plt.axhline(y = mean(x) + stdev(x), color='r', linestyle='--')
plt.axhline(y = mean(x) - stdev(x), color='r', linestyle='--')
plt.title('MC Crudo')
plt.grid(True)
plt.show()
```
