```python
from random import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función que devuelve la proporción de veces que que el tiempo de espera es
# menor o igual a 10 minutos
def Tiempo(m):
  if m != 0:
    c = 0
    for i in range(m):
      t = 30*abs(random() - random())
      if t <= 10:
        c += 1
    return c/m
```

```python
# Iteración sobre 1000 repeticiones del experimento
tiempo = []
n = 1000
for i in range(n):
  tiempo.append(Tiempo(i + 1))

mu = np.mean(tiempo)
sigma = np.std(tiempo)
E = 1.96*np.sqrt(mu*(1 - mu)/n)

plt.plot(tiempo, linewidth = 0.4)
plt.axhline(mu, color = "r", linestyle = "--", label = f"Media: {mu:.4f}")
plt.fill_between(range(0, n), mu - E, mu + E, color = "r", alpha = 0.2, label = f"IC al 95%: [{mu - E:.4f}, {mu + E:.4f}]")
plt.title(f"Simulación con {n} iteraciones")
plt.xlabel("Iteración")
plt.ylabel("P(t <= 10)")
plt.legend(loc = "best")
plt.show()
```

# Comparación de resultados
La solución analítica nos indica que la probabilidad de que el tiempo de espera sea menor o igual a 10 minutos es del 55.56%. Por otro lado, el modelo de simulación nos da una probabilidad estimada del 55.41%, con un intervalo de confianza del 95% que abarca aproximadamente ± 3.08%. Dado que la probabilidad teórica se encuentra dentro de este intervalo, podemos concluir que la simulación es consistente y precisa, y valida adecuadamente el resultado analítico.
