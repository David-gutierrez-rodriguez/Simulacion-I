```python
import pandas as pd
```

```python
def rand_num(seed):
    """
    Genera una secuencia de números pseudoaleatorios utilizando el método de los cuadrados medios.

    Parámetros:
    seed (int): Valor inicial (semilla).

    Retorna:
    dict: Diccionario con la secuencia generada, incluyendo el número original, su cuadrado y su versión normalizada.
    """
    rand = {"x": [seed], "xˆ2": [seed**2], "x/10000": [seed / 10000]}  # Inicializa diccionario con la semilla

    # Calcula el primer número pseudoaleatorio
    xn = seed ** 2                # Eleva al cuadrado la semilla
    xn = xn // 100                # Desplaza dos dígitos a la derecha
    xn -= (xn // 10000) * 10000   # Toma solo los 4 dígitos centrales

    i = 1  # Contador de iteraciones

    # Genera números hasta que se repita un valor
    while xn not in rand["x"]:
        rand["x"].append(xn)
        rand["xˆ2"].append(xn ** 2)
        rand["x/10000"].append(xn / 10000)

        xn = rand["xˆ2"][i]           # Eleva al cuadrado el nuevo valor
        xn = xn // 100                # Desplaza dos dígitos a la derecha
        xn -= (xn // 10000) * 10000   # Toma solo los 4 dígitos centrales

        i += 1  # Incrementa el contador

    return rand
```

```python
# Generación de números pseudoaleatorios con la semilla dada
randnum = rand_num(6010)
print(f"Ciclo de vida: {len(randnum['x'])}")  # Imprime la longitud del ciclo antes de repetirse
```
