```python
def KS(nums):
    """
    Calcula las estadísticas necesarias para la prueba de Kolmogorov-Smirnov.

    Parámetros:
    - nums: lista de números aleatorios entre 0 y 1.

    Retorna:
    - Un diccionario con los valores ordenados (x), D+ y D-.
    """

    # Ordenamos los números de menor a mayor
    nums.sort()
    n = len(nums)

    # Listas para almacenar las diferencias D+ y D-
    D1, D2 = [], []

    # Calculamos D+ y D-
    for i in range(1, n + 1):
        D1.append(abs(i / n - nums[i - 1]))         # Diferencia con la CDF teórica
        D2.append(abs(nums[i - 1] - (i - 1) / n))   # Diferencia con la CDF empírica

    # Retornamos los valores calculados
    return {"x": nums, "D+": D1, "D-": D2}
```

```python
# Lista de números aleatorios a evaluar
numeros = [0.048, 0.926, 0.368, 0.787, 0.298, 0.591, 0.007, 0.437, 0.594, 0.199, 0.478, 0.551,
           0.178, 0.086, 0.157, 0.166, 0.362, 0.411, 0.951, 0.401, 0.932, 0.533, 0.545, 0.772, 0.631]
```

```python
import pandas as pd

# Convertimos el resultado en un DataFrame para mejor visualización
nums = pd.DataFrame(KS(numeros))
display(nums)
```

```python
# Calculamos los valores máximos de D+ y D-
D1 = max(nums["D+"])
D2 = max(nums["D-"])
D = max(D1, D2)  # Estadístico D de Kolmogorov-Smirnov

# Mostramos los valores calculados
print(f"D+ = {D1}")
print(f"D- = {D2}")
print(f"D = {D}\n")
```

```python
from scipy.stats import ksone

# Nivel de significancia
alpha = 0.05
n = len(numeros)

# Calculamos el valor crítico usando la distribución de KS
d_a_n = ksone.ppf(1 - alpha / 2, n)

# Mostramos el valor crítico
print(f"Valor crítico: {d_a_n}")

# Comparamos D con el valor crítico para determinar si la muestra es uniforme
if D > d_a_n:
    print("Muestra no uniforme")
else:
    print("Muestra uniforme")
```
